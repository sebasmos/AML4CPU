```python
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerBase

model_info = {
    'XGBoost Regressor': {'color': "skyblue", 'marker': "o"},
    'Random Forest': {'color': "salmon", 'marker': "s"},
    'Linear Regression': {'color': "gold", 'marker': "^"},
    'Ada Boost Regressor': {'color': "lightgreen", 'marker': "D"},
    'Decision Tree Regressor': {'color': "lightcoral", 'marker': "p"},
    'MLP Regressor': {'color': "plum", 'marker': "X"},
    'Random Forest Regressor': {'color': "lightblue", 'marker': "o"},
    'Passive Aggressive Regressor': {'color': "lightseagreen", 'marker': "s"},
    'K-Neighbors Regressor': {'color': "lightpink", 'marker': "^"},
    'SVR': {'color': "blue", 'marker': "D"},
    'Linear Regression (LR)': {'color': "lightcoral", 'marker': "p"},
    'LSTM with Attention': {'color': "lightgreen", 'marker': "X"},
    'GRU': {'color': "grey", 'marker': "o"},
    'BI-LSTM': {'color': "lightsalmon", 'marker': "s"},
    'LSTM': {'color': "red", 'marker': "^"},
    'SGD Regressor': {'color': "lightseagreen", 'marker': "D"}
}

scatter_handles = {}  # Dictionary to store scatter plot handles with labels

fig, axs = plt.subplots(2, 3, figsize=(12, 8), sharey=True)  # 2 rows, 3 columns for 6 window sizes

class MarkerHandler(HandlerBase):
    def create_artists(self, legend, tup, xdescent, ydescent,
                       width, height, fontsize, trans):
        return [tup]

for i, (col, ax) in enumerate(zip(df["Window Size"], axs.flatten())):

    df_temp = df[df['Window Size'] == col].reset_index(drop=True)
    categories = np.random.choice(len(model_info), len(df_temp))
    print(f"There are {len(df_temp)} points out of {len(df)}")
    for index, row in df_temp.iterrows():
        model_info_entry = model_info[row['Model']]
        color = model_info_entry['color']
        marker = model_info_entry['marker']
        
        print(f"Model name: {row.Model} c/olor: {color}, Marker: {marker}")
        
        handle = ax.scatter(row['MAE_mean'], row['Training_time'], color=color, marker=marker, label=row['Model'], s=50)
        if i == 0:  
            scatter_handles[row['Model']] = handle

    ax.set_title(f"Window Size {col}") 
    ax.set_yscale('log') 
    # ax.set_xscale('log') 
    ax.set_xlabel('MAE')
    ax.set_ylabel('Inference Time (log scale)')
    ax.grid(True)  

plt.subplots_adjust(right=0.85)
# Create legend based on scatter_handles
fig.legend(handles=scatter_handles.values(), labels=scatter_handles.keys(), loc='center left', bbox_to_anchor=(1, 0.5), ncol=1, handler_map={plt.scatter: MarkerHandler()})
plt.tight_layout()

plt.show()
```

```python

from matplotlib.ticker import ScalarFormatter


fig, axs = plt.subplots(2, 3, figsize=(14, 8), sharey=True)  # Increase figure width to 14

plt.subplots_adjust(hspace=0.5)  # Increase vertical spacing between subplots

class MarkerHandler(HandlerBase):
    def create_artists(self, legend, tup, xdescent, ydescent,
                       width, height, fontsize, trans):
        return [tup]

for i, (col, ax) in enumerate(zip(df["Window Size"], axs.flatten())):

    df_temp = df[df['Window Size'] == col].reset_index(drop=True)
    categories = np.random.choice(len(model_info), len(df_temp))
    print(f"There are {len(df_temp)} points out of {len(df)}")
    for index, row in df_temp.iterrows():
        model_info_entry = model_info[row['Model']]
        color = model_info_entry['color']
        marker = model_info_entry['marker']
        
        print(f"Model name: {row.Model} c/olor: {color}, Marker: {marker}")
        
        handle = ax.scatter(row['MAE_mean'], row['Training_time'], color=color, marker=marker, label=row['Model'], s=50)
        if i == 0:  
            scatter_handles[row['Model']] = handle

    ax.set_title(f"Window Size {col}") 
    ax.set_yscale('log') 
    # ax.set_xscale('log') 
    ax.set_xlabel('MAE')
    ax.set_ylabel('Inference Time (log scale)')
    
    # ax.set_xticks(x[1:]) # note that with a log axis, you can't have x = 0 so that value isn't plotted.
    # ax.xaxis.set_major_formatter(ScalarFormatter())
    
    x = [0, 1, 2,3, 4,5, 6.5, 8 , 10 ,15,20,25]
    # ax.set_xscale('log')
    ax.set_xticks(x)
    ax.set_xticklabels(x)
    ax.set_xlim([0,30])
    ax.grid(True)  

plt.subplots_adjust(right=1)


# Create legend based on scatter_handles
fig.legend(handles=scatter_handles.values(), labels=scatter_handles.keys(), loc='center left', bbox_to_anchor=(1, 0.5), ncol=1, handler_map={plt.scatter: MarkerHandler()})
plt.tight_layout()
plt.show()
```
